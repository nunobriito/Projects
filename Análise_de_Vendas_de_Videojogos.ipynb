# **Qual foi a plataforma com mais vendas  para cada gênero de videojogo?**

Cada gênero pode ser mais popular em determinadas plataformas, e é interessante investigar qual foi a plataforma que teve mais vendas para cada um deles. Com este gráfico, poderemos ter uma visão geral de qual foi a plataforma mais bem-sucedida para cada gênero de videojogo.

```python
# Agrupar os dados de acordo com o gênero e calcular a soma das vendas de cada plataforma para cada gênero
vendas_por_genero_plataforma = df.groupby(['Gênero', 'Plataforma'])['Vendas_Globais'].sum()
top10_vendas_por_genero_plataforma = vendas_por_genero_plataforma.sort_values(ascending=False).head(20)

# Criar um gráfico de barras com os dados agrupados
top10_vendas_por_genero_plataforma.plot.bar(figsize=(20,12),color="#2d6cdf")

# Adicionar títulos e legendas
plt.title('Plataforma mais popular para cada gênero', fontsize=20)
plt.xlabel('Gênero e Plataforma', fontsize=14)
plt.ylabel('Vendas (em milhões de dólares)', fontsize=14)

# Adicionar valores nas barras
for barra in plt.gca().patches:
    plt.text(barra.get_x() + barra.get_width() / 2, 
             barra.get_height() + 0.5,
             f"{barra.get_height()}", 
             ha='center', 
             va='bottom',
             fontsize=10)

# Remover a linha exterior
ax.set_frame_on(False)  
    
# Mostrar o gráfico
plt.show()
```

No gráfico acima conseguimos ver o top 20 de vendas por gênero em cada plataforma. Neste caso a plataforma e gênero com mais faturamento foram na PS3 com videojogos de Ação. 
Uma vez que entramos em vendas por plataforma será interessante também perceber qual foi o faturamento por plataforma globalmente.

```python
#Agrupar valores de vendas com plataforma e somar
vendas_por_plataforma = df.groupby("Plataforma")["Vendas_Globais"].sum().sort_values(ascending=False)

# Criar um gráfico de barras com os dados agrupados
vendas_por_plataforma.plot.bar(figsize=(20,12),color="#2d6cdf")

#Adicionar titulos e legendas
plt.title("Vendas por Plataforma", fontsize=20)
plt.xlabel("Plataformas",fontsize=14)
plt.ylabel("Vendas (em milhões de doláres)",fontsize=14)
# Remover a linha exterior
ax.set_frame_on(False)

#Mostrar o gráfico
plt.show()
```

Este gráfico mostra as plataformas com as mais vendas para cada gênero de videojogo. É evidente que a plataforma **PS2** foi a que mais vendeu em todos os gêneros, com diferenças de **mais de 200 milhões de dólares** em relação à segunda colocada, a **Xbox 360**. Também observamos uma diferença significativa entre a sexta colocada, a **PS**, e a sétima, a **Game Boy Advance (GBA)**, com quase 50% de diferença.

É possível que essas diferenças de faturamento possam estar relacionadas com o número de jogos lançados para cada plataforma. Pode ser que a PS2 tenha tido um número maior de jogos lançados, o que explicaria a sua liderança em vendas. Seria interessante analisar mais detalhadamente esses dados para ter uma compreensão mais profunda das razões por trás dessas diferenças de faturamento.

```python
# Agrupar dados por plataforma e fazer a soma dos jogos
df.groupby("Plataforma")["Nome"].count().sort_values(ascending=False).head(10)
```

O faturamento recorde da **PS2** pode ser explicado pelo grande número de jogos lançados para essa plataforma. Em comparação com a **Nintendo DS**, que lançou 2163 jogos, a **PS2** lançou apenas 2 jogos a menos, com 2161. No entanto, essa diferença é muito maior em relação às outras plataformas, como a **PS3**, que lançou 1329 jogos e ficou em terceiro lugar em termos de faturamento.

Para entender melhor essa questão, vamos dividir o faturamento pelo número de jogos lançados em cada plataforma, o que nos dará uma média do faturamento por jogo em cada uma delas.

```python
# Fazer média de faturamento por plataforma
numero_jogos_por_plataforma = df.groupby("Plataforma")["Nome"].count()
faturamento_medio_por_plataforma = vendas_por_plataforma / numero_jogos_por_plataforma
faturamento_medio_por_plataforma.sort_values(ascending=False)
```

A **PS2** foi a plataforma de videojogos com o maior faturamento, graças ao grande número de jogos lançados para ela. Em comparação com a **PS3**, que teve um faturamento menor mas lançou menos jogos, a média de faturamento por jogo da **PS2** foi de **580 mil dólares**, enquanto a média da **PS3 foi de 720 mil dólares**. Isso sugere que, se a **PS3** tivesse lançado o mesmo número de jogos que a **PS2**, o seu faturamento poderia ter sido de cerca de 1555 milhões de dólares, ou seja, mais 300 mil dólares do que a **PS2**.

* O Japão tem sido uma região importante para o mercado de jogos eletrônicos, com receitas significativamente maiores do que as da Europa e da América do Norte em certos anos, especificamente em 1985 e 1991, com o lançamento do **NES (Nintendo Entertainment System)** e **SNES (Super Nintendo Entertainment System)** respectivamente. Com essas consolas vieram alguns dos mais populares jogos do mundo como:  **Contra**, **The Legend of Zelda: A Link to the Past**,  **Super Mario World** e **Street Fighter 2**. 

* O gênero de jogos de ação foi o mais popular, seguido pelo desporto e jogos de tiro.

* A plataforma **PS2** foi a mais vendida desde 1980 e a **PS2, X360 e PS3** foram as plataformas mais vendidas desde 2000.

* A **Nintendo** foi a editora com as maiores vendas no Japão e no mundo, seguida de perto pela **Electronic Arts**, enquanto a **Namco Bandai Games** teve uma posição de destaque nas vendas no Japão.

* Analisando o gráfico de barras cruzadas dos gêneros e plataformas, podemos ver que cada gênero tem uma plataforma que se destaca como a mais popular para esse gênero específico. Por exemplo, jogos de plataforma exclusivos são mais populares nas plataformas da **Playstation** (**PS2, PS3 e PS4**) e na **X360**, enquanto jogos de simulação são mais populares na **Nintendo DS** e no **PC**. 

* O PC tem tido um crescimento considerável nas vendas desde 2000, isso pode ser atribuído ao aumento do uso de computadores e a popularidade de jogos online.

Além dessas descobertas, seria importante incluir recomendações para as empresas e indústria de jogos baseadas nos insights encontrados, como por exemplo, as editoras de jogos devem continuar a investir em jogos de ação e jogos de esportes, pois esses gêneros têm a maior demanda entre os consumidores e as plataformas devem continuar a investir em jogos de plataforma exclusivos para atrair a atenção dos consumidores, uma vez que as vendas de plataforma exclusiva têm uma tendência de baixa no mercado de videojogos. As tendências de vendas globais mostraram que as preferências dos consumidores mudam com o tempo e é importante estar atento a essas mudanças para se adaptar e oferecer jogos que os consumidores estão procurando.
